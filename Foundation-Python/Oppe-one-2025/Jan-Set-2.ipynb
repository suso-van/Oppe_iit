```python
def is_both_divisible_by(a: int, b: int, c: int) -> bool:
    """
    Checks if the integer `c` divides both integers `a` and `b`.

    Args:
        a (int): The first integer.
        b (int): The second integer.
        c (int): The divisor (assumed to be non-zero).

    Returns:
        bool: True if `c` divides both `a` and `b`, False otherwise.
    """
    return a % c == 0 and b % c == 0
```

```python
print(f"is_both_divisible_by(10, 20, 5): {is_both_divisible_by(10, 20, 5)}")
print(f"is_both_divisible_by(10, 21, 5): {is_both_divisible_by(10, 21, 5)}")
print(f"is_both_divisible_by(12, 18, 6): {is_both_divisible_by(12, 18, 6)}")
print(f"is_both_divisible_by(7, 14, 3): {is_both_divisible_by(7, 14, 3)}")
```

SECTION ONE PROBLEM - 2

```python
def is_rotation(str1: str, str2: str) -> bool:
    """
    Checks if one string is a rotation of another.

    Example:
      >>>is_rotation('abcde','cdeab')
      True
      >>>is_rotation('hello','ehllo')
      False
      >>>is_rotation('hello','helol')
      False

    Args:
        str1 (str): The original string.
        str2 (str): The string to check if it's a rotation of str1.

    Returns:
        bool: True if str2 is a rotation of str1, False otherwise.
    """
    if len(str1) != len(str2):
        return False
    if not str1: # Handles empty strings, or if both are empty
        return True

    for i in range(len(str1)):
        # Generate a rotation by splitting str1 at index i and swapping the parts
        rotated_str1 = str1[i:] + str1[:i]
        if rotated_str1 == str2:
            return True
    return False
```

```python
print(f"is_rotation('abcde','cdeab'): {is_rotation('abcde','cdeab')}")
print(f"is_rotation('hello','ehllo'): {is_rotation('hello','ehllo')}")
print(f"is_rotation('hello','helol'): {is_rotation('hello','helol')}")
print(f"is_rotation('waterbottle', 'erbottlewat'): {is_rotation('waterbottle', 'erbottlewat')}")
print(f"is_rotation('apple', 'pleap'): {is_rotation('apple', 'pleap')}")
print(f"is_rotation('camera', 'macera'): {is_rotation('camera', 'macera')}")
print(f"is_rotation('', ''): {is_rotation('', '')}")
```

```python
def find_missing_number(nums: list, n: int) -> int:
    """Finds the missing number in a list of integers ranging from 1 to n.

    Args:
        nums (list): A list of integers from 1 to n with one number missing.
        n (int): The number n describing the range
    Returns:
        int: The missing number in the list.
    """
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum
```

```python
print(f"find_missing_number([1, 2, 4, 5, 6], 6): {find_missing_number([1, 2, 4, 5, 6], 6)}")
print(f"find_missing_number([1, 3], 3): {find_missing_number([1, 3], 3)}")
print(f"find_missing_number([2, 3, 4, 5], 5): {find_missing_number([2, 3, 4, 5], 5)}")
print(f"find_missing_number([], 1): {find_missing_number([], 1)}")
```

```python
def merge_dictionaries(d1:dict, d2:dict):
    """
    Merges two dictionaries by summing values of common keys.

    If a key exists in both `d1` and `d2`, their values are summed.
    Keys that are unique to one dictionary are included as-is in the result.

    Args:
        d1 (dict): The first dictionary, where values are integers or floats.
        d2 (dict): The second dictionary, where values are integers or floats.

    Returns:
        dict: A new dictionary containing all keys from both `d1` and `d2`.
        Values for overlapping keys are summed, while unique keys retain their original values.

    """
    merged_dict = d1.copy()
    for key, value in d2.items():
        merged_dict[key] = merged_dict.get(key, 0) + value
    return merged_dict
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 5, 'c': 7, 'd': 8}
print(f"merge_dictionaries({dict1}, {dict2}): {merge_dictionaries(dict1, dict2)}")

dict3 = {'x': 10, 'y': 20}
dict4 = {'y': 5, 'z': 15}
print(f"merge_dictionaries({dict3}, {dict4}): {merge_dictionaries(dict3, dict4)}")

dict5 = {'apple': 1.5, 'banana': 0.75}
dict6 = {'banana': 0.25, 'orange': 2.0}
print(f"merge_dictionaries({dict5}, {dict6}): {merge_dictionaries(dict5, dict6)}")

dict7 = {}
dict8 = {'a': 1}
print(f"merge_dictionaries({dict7}, {dict8}): {merge_dictionaries(dict7, dict8)}")

dict9 = {'a': 1}
dict10 = {}
print(f"merge_dictionaries({dict9}, {dict10}): {merge_dictionaries(dict9, dict10)}")
```

```python
from collections import Counter

def find_most_frequent_numbers():
    """
    Reads n numbers from input, finds the most frequent numbers,
    and prints them in ascending order.
    """
    n = int(input())
    numbers = []
    for _ in range(n):
        numbers.append(int(input()))

    if not numbers:
        print([])
        return

    # Count frequencies of each number
    counts = Counter(numbers)

    # Find the maximum frequency
    max_frequency = 0
    if counts:
        max_frequency = max(counts.values())

    # Collect all numbers that have the maximum frequency
    most_frequent = []
    for num, freq in counts.items():
        if freq == max_frequency:
            most_frequent.append(num)

    # Sort the most frequent numbers in ascending order
    most_frequent.sort()

    print(most_frequent)

# To run the function with the example input, uncomment the line below:
# find_most_frequent_numbers()
```

```python
def employees_with_salary_above(employees:list, min_salary:int):
    """Returns the employees names with salary greater than are equal to the given salary.

    Args:
        employees (list[dict]):
            list of dictionary of employees with
            the keys name, salary and department.
        min_salary (int): The cutoff salary.

    Returns:
        list[str]:
            list of names of the employees with salary
            greater than or equal to the cutoff
            in the order of occurance.
    """
    result = []
    for employee in employees:
        if employee['salary'] >= min_salary:
            result.append(employee['name'])
    return result


def total_salary_in_department(employees:list, department:str):
    """Returns the total salary of all the employees in the department.

    Args:
        employees (list[dict]):
            list of dictionary of employees with
            the keys name, salary and department.
        department (str): The department to find the total salary.

    Returns:
        int: The total salary of the employees in the given department.
    """
    total = 0
    for employee in employees:
        if employee['department'] == department:
            total += employee['salary']
    return total


def ceil_to_five_hundreds(num: int):
    """Given an integer, increase it to the next multiple of 500 if it is not a multiple of 500.

    Args:
        num (int): The number to increment.

    Returns:
        int: The number ceiled to the next five hundreds.

    Examples:
    >>> ceil_to_five_hundreds(24500)
    24500
    >>> ceil_to_five_hundreds(24600)
    25000
    >>> ceil_to_five_hundreds(24400)
    24500
    """
    if num % 500 == 0:
        return num
    else:
        return (num // 500 + 1) * 500



def max_salary_after_increment_in_department(employees:list, department:str, inc_percent:int):
    """Returns the maximum salary in the given department after increment.

    The Maximum salary is incremented with the given percentage,
    rounded to the nearest integer and ceiled to the next five hundreds

    Rounding can be done using round function.

    Args:
        employees (list[dict]):
            list of dictionary of employees with
            the keys name, salary and department.
        department (str): The department to find the maximum salary.
        inc_percent (int): Percentage of increment in the salary.

    Returns:
        int:
            The maximum salary in the given department after increment
            and ceiling it to the next five hundreds.
    """
    max_sal = 0
    for employee in employees:
        if employee['department'] == department:
            if employee['salary'] > max_sal:
                max_sal = employee['salary']

    if max_sal == 0:
        return 0 # No employees in the department or all have 0 salary

    incremented_salary = max_sal * (1 + inc_percent / 100)
    rounded_salary = round(incremented_salary)
    ceiled_salary = ceil_to_five_hundreds(rounded_salary)
    return ceiled_salary
```

```python
employees_data = [
    {'name': 'Alice', 'salary': 50000, 'department': 'HR'},
    {'name': 'Bob', 'salary': 60000, 'department': 'Engineering'},
    {'name': 'Charlie', 'salary': 75000, 'department': 'HR'},
    {'name': 'David', 'salary': 80000, 'department': 'Engineering'},
    {'name': 'Eve', 'salary': 55000, 'department': 'HR'}
]

# Test employees_with_salary_above
print(f"Employees with salary above 55000: {employees_with_salary_above(employees_data, 55000)}")
print(f"Employees with salary above 70000: {employees_with_salary_above(employees_data, 70000)}")
print(f"Employees with salary above 90000: {employees_with_salary_above(employees_data, 90000)}")

# Test total_salary_in_department
print(f"Total salary in HR department: {total_salary_in_department(employees_data, 'HR')}")
print(f"Total salary in Engineering department: {total_salary_in_department(employees_data, 'Engineering')}")
print(f"Total salary in Sales department: {total_salary_in_department(employees_data, 'Sales')}")

# Test ceil_to_five_hundreds
print(f"Ceil 24500 to five hundreds: {ceil_to_five_hundreds(24500)}")
print(f"Ceil 24600 to five hundreds: {ceil_to_five_hundreds(24600)}")
print(f"Ceil 24400 to five hundreds: {ceil_to_five_hundreds(24400)}")
print(f"Ceil 0 to five hundreds: {ceil_to_five_hundreds(0)}")

# Test max_salary_after_increment_in_department
print(f"Max salary in HR after 10% increment: {max_salary_after_increment_in_department(employees_data, 'HR', 10)}")
print(f"Max salary in Engineering after 5% increment: {max_salary_after_increment_in_department(employees_data, 'Engineering', 5)}")
print(f"Max salary in Marketing (non-existent) after 10% increment: {max_salary_after_increment_in_department(employees_data, 'Marketing', 10)}")
```

```python
def draw_z_pattern(n: int):
    """
    Draws a Z-shaped pattern using asterisks (*).

    The pattern consists of:
    - A row of asterisks at the top.
    - A diagonal line of asterisks from the top-right to the bottom-left.
    - A row of asterisks at the bottom.

    Args:
        n (int): The size of the Z pattern (number of rows and columns). Guaranteed n >= 3.
    """
    # Top row
    print('*' * n)

    # Diagonal rows
    # The diagonal asterisk is at column `n - 1 - row_index`
    # `row_index` ranges from 1 to n-2 (exclusive of top and bottom rows)
    for i in range(1, n - 1):
        row_chars = [' '] * n
        diagonal_pos = n - 1 - i
        row_chars[diagonal_pos] = '*'
        print(''.join(row_chars))

    # Bottom row
    print('*' * n)

# Example usage (uncomment to test with a specific N):
# n_input = int(input("Enter the size of the Z pattern (n): "))
# draw_z_pattern(n_input)
```

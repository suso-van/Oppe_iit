```python
def sum_of_floored_to_tens(a:int, b:int):
    """
    Calculate the sum of two integers floored to the nearest lower multiple of 10.

    Args:
        a (int): The first integer.
        b (int): The second integer.

    Returns:
        int: The sum of the two integers after flooring each to the nearest lower multiple of 10.

    Examples:
        >>> sum_of_floored_to_tens(23, 47)
        60
        >>> sum_of_floored_to_tens(35, 19)
        40
        >>> sum_of_floored_to_tens(0, 105)
        100
    """


    a = a // 10 * 10
    b = b // 10 * 10
    return a + b

if __name__ == '__main__':
    print(f"sum_of_floored_to_tens(23, 47) -> {sum_of_floored_to_tens(23, 47)}")
    print(f"sum_of_floored_to_tens(35, 19) -> {sum_of_floored_to_tens(35, 19)}")
    print(f"sum_of_floored_to_tens(0, 105) -> {sum_of_floored_to_tens(0, 105)}")
```

```python
def surround_first_two_and_last_two_with_brackets(s: str):
    """
    Surrounds the first two and last two characters of the input string with square brackets.

    It is assumed that the input has atleast four characters.

    Args:
        s (str): The input string.

    Returns:
        str: A new string with the first two and last two characters surrounded by square brackets.

    Example:
        >>> surround_first_two_and_last_two_with_brackets("example")
        '[ex]amp[le]'
    """

    return f"[{s[:2]}]{s[2:-2]}[{s[-2:]}]"

if __name__ == '__main__':
    print(f'surround_first_two_and_last_two_with_brackets("example") -> {surround_first_two_and_last_two_with_brackets("example")}')
    print(f'surround_first_two_and_last_two_with_brackets("python") -> {surround_first_two_and_last_two_with_brackets("python")}')
```

```python
def number_with_more_unique_digits(n1:int, n2:int):
    """
    Compares two integers and returns the one with more unique digits.

    Args:
        n1 (int): The first integer.
        n2 (int): The second integer.

    Returns:
        int or tuple: The integer with more unique digits.
        If both numbers have the same number of unique digits,
        returns a tuple of both integers `(n1, n2)`.

    Example:
        >>> number_with_more_unique_digits(123, 1122)
        123

        >>> number_with_more_unique_digits(1122, 2211)
        (1122, 2211)
    """


    n1_n_unique = len(set(str(n1)))
    n2_n_unique = len(set(str(n2)))
    if n1_n_unique> n2_n_unique:
        return n1
    elif n2_n_unique> n1_n_unique:
        return n2
    else:
        return (n1,n2)

if __name__ == '__main__':
    print(f"number_with_more_unique_digits(123, 1122) -> {number_with_more_unique_digits(123, 1122)}")
    print(f"number_with_more_unique_digits(1122, 2211) -> {number_with_more_unique_digits(1122, 2211)}")
```

```python
def count_vowels_and_consonants_in_even_indices(s: str) -> tuple:
    """
    Counts the number of vowels and consonants at even indices in the given string.

    Args:
        s (str): The input string.

    Returns:
        tuple: A tuple containing two integers:
            - The first integer is the count of vowels at even indices.
            - The second integer is the count of consonants at even indices.

    Example:
        >>> count_vowels_and_consonants_in_even_indices("example1")
        (3, 1)

        >>> count_vowels_and_consonants_in_even_indices("hello")
        (1, 2)
    """


    vowel_count = 0
    consonant_count = 0
    for c in s[::2].lower():
        if c in "aeiou":
            vowel_count += 1
        elif c.isalpha():
            consonant_count += 1
    return vowel_count, consonant_count

if __name__ == '__main__':
    print(f"'example1': {count_vowels_and_consonants_in_even_indices('example1')}")
    print(f"'hello': {count_vowels_and_consonants_in_even_indices('hello')}")
```

```python
def get_call_counts(call_logs):
    call_counts = {}
    for log in call_logs:
        if log['name'] not in call_counts:
            call_counts[log['name']] = 0
        call_counts[log['name']] += 1
    return call_counts

def get_total_call_durations(call_logs):
    total_durations = {}
    for log in call_logs:
        if log['name'] not in total_durations:
            total_durations[log['name']] = 0
        total_durations[log['name']] += log['duration']
    return total_durations

def most_frequent_caller(call_logs):
    call_counts = get_call_counts(call_logs)
    total_durations = get_total_call_durations(call_logs)
    return max(
        call_counts,
        key=lambda name: (call_counts[name], total_durations[name])
    )


def process_call_logs(call_logs, task):


        if task == 'get_call_counts':
            return get_call_counts(call_logs)
        elif task == 'get_total_call_durations':
            return get_total_call_durations(call_logs)
        elif task == 'most_frequent_caller':
            return most_frequent_caller(call_logs)
        else:
            raise ValueError("Invalid task provided.")

# Example usage to demonstrate the functions
if __name__ == '__main__':
    sample_call_logs = [
        {'name': 'Alice', 'duration': 10},
        {'name': 'Bob', 'duration': 15},
        {'name': 'Alice', 'duration': 20},
        {'name': 'Charlie', 'duration': 5},
        {'name': 'Bob', 'duration': 10},
        {'name': 'Alice', 'duration': 5}
    ]

    print("--- Call Counts ---")
    call_counts_result = process_call_logs(sample_call_logs, 'get_call_counts')
    print(call_counts_result)

    print("\n--- Total Call Durations ---")
    total_durations_result = process_call_logs(sample_call_logs, 'get_total_call_durations')
    print(total_durations_result)

    print("\n--- Most Frequent Caller ---")
    most_frequent_result = process_call_logs(sample_call_logs, 'most_frequent_caller')
    print(most_frequent_result)
```

```python
# This writes the stdin to the input file
import tempfile
import sys
import io

# Temporarily redirect sys.stdin to provide sample input
original_stdin = sys.stdin
sys.stdin = io.StringIO("First Line\nSecond Line\nThird Line\n2,1,3")

_, filename = tempfile.mkstemp(prefix="case")
with open(filename, 'w') as f:
    f.write(sys.stdin.read())

# Restore original sys.stdin
sys.stdin = original_stdin


with open(filename) as f:
    lines = f.readlines()
    lines, ordering = lines[:-1], lines[-1].strip().split(",")
    ordered_lines = [lines[int(num) - 1] for num in ordering]
    print(*ordered_lines, sep="")
```

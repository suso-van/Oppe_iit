```
def sum_squares_abs_diff_squares() -> tuple:
    '''Return the sum of squares and the absolute difference of squares.

    Args:
        a, b : int - input numbers (Note: These are now obtained internally via input())

    Returns:
        tuple - tuple with sum of squares and absolute difference of squares.

    '''
    a = int(input("Enter the first number: "))
    b = int(input("Enter the second number: "))
    sum_squares = a**2 + b**2
    abs_diff_squares = abs(a**2 - b**2)
    print(tuple([sum_squares, abs_diff_squares]))
sum_squares_abs_diff_squares()
```

```
def is_odd_length_palindrome() -> bool:
    '''Check if a string is a palindrome with odd length.

    Args:
        s : str - input string (Note: This is now obtained internally via input())

    Returns:
        bool - True if s is a palindrome with odd length, False otherwise.
    '''
    s = str(input("Enter a string: "))
    if len(s) % 2 == 1 and s == s[::-1]:
      return True
    else:
      return False
is_odd_length_palindrome()
```

```
def remove_elements_at_two_indices():
    '''Remove elements at two specified indices in the list.

    Args:
        l : list - input list (Note: This is now obtained internally via input())
        i1, i2 : int - indices of elements to remove (Note: These are now obtained internally via input())
            both are non-negative and unique

    Returns:
        None - modifies the list in place
    '''
    l_str = input("Enter the list (e.g., [1, 2, 3]): ")
    l = eval(l_str) # Using eval for simple list input, but safer methods might be preferred for production
    i1 = int(input("Enter the first index: "))
    i2 = int(input("Enter the second index: "))

    # Ensure indices are valid and in correct order for popping
    if i1 > i2:
        i1, i2 = i2, i1

    # Validate indices before popping
    if i2 >= len(l) or i1 < 0:
        print(f"Error: One or both indices ({i1}, {i2}) are out of range for a list of length {len(l)}. No elements were removed.")
        return

    l.pop(i2)
    l.pop(i1)
    print(l)
remove_elements_at_two_indices()
```

```
def sum_of_squares_of_even(nums: list) -> int:
    '''Return the sum of squares of all even numbers in the list.

    Args:
        nums : list - list of integers

    Returns:
        int - sum of squares of all even numbers
    '''
    total_sum = 0
    for num in nums:
        if num % 2 == 0:
            total_sum += num ** 2
    return total_sum

# Example usage:
print(sum_of_squares_of_even([1, 2, 3, 4, 5, 6])) # Expected: 2*2 + 4*4 + 6*6 = 4 + 16 + 36 = 56
```

```
''''
Upper Case Only Vowels
Write a program that takes a passage with n lines of text as input. For each line, convert all vowels (a, e, i, o, u) to uppercase and all other characters to lowercase.

NOTE: This is an I/O type question, you need to write the whole code for taking input and printing the output.

Input Format

    • The first line contains an integer n, the number of lines.

    • The next n lines contain the passage.

Output Format

    • Print the passage with vowels in uppercase and all other characters in lowercase.

Hint
Use the upper and lower methods of str to convert characters as needed.

Example

Input:

2
Hello World
Python Programming
Output:

hEllO wOrld
pythOn prOgrAmmIng
'''

n = int(input())
vowels = 'aeiou'

for _ in range(n):
    line = input()
    result_line = []
    for char in line:
        if char.lower() in vowels:
            result_line.append(char.upper())
        else:
            result_line.append(char.lower())
    print(''.join(result_line))
```

```
def get_roll_nos(data: list, criteria=None) -> list:
    '''
    Filter roll numbers based on criteria.

    Args:
        data (list): List of tuples with roll number and marks.
        criteria (str, optional): The criteria for filtering.

    Returns:
        list: List of roll numbers matching the criteria or None if invalid criteria.
    '''
    if not data:
        return [] if criteria in ['above_average', 'below_average', 'fail', 'toppers', None] else None

    roll_numbers = []
    marks = [item[1] for item in data]

    if criteria is None:
        return [item[0] for item in data]
    elif criteria == 'above_average' or criteria == 'below_average':
        if not marks:
            return []
        average_marks = sum(marks) / len(marks)
        if criteria == 'above_average':
            for roll_no, mark in data:
                if mark >= average_marks:
                    roll_numbers.append(roll_no)
        else: # 'below_average'
            for roll_no, mark in data:
                if mark < average_marks:
                    roll_numbers.append(roll_no)
    elif criteria == 'fail':
        for roll_no, mark in data:
            if mark < 40:
                roll_numbers.append(roll_no)
    elif criteria == 'toppers':
        for roll_no, mark in data:
            if mark >= 90:
                roll_numbers.append(roll_no)
    else:
        return None # Invalid criteria

    return roll_numbers

# Example usage from problem description (for testing purposes, not part of function definition)
# data = [("101", 85), ("102", 75), ("103", 45), ("104", 95), ("105", 35)]
# print(get_roll_nos(data, 'above_average'))
# print(get_roll_nos(data, 'fail'))
# print(get_roll_nos(data, 'toppers'))
# print(get_roll_nos(data))
# print(get_roll_nos(data, 'invalid_criteria'))
# print(get_roll_nos([], 'above_average')) # Test with empty data
```

```
'''
Print Pattern - V
Given an integer n (where n > 0), print a "V" shaped pattern with n rows. The pattern should use backslashes (\) and forward slashes (/) for each row, with the v character at the bottom. There should be no spaces to the right of the pattern.

NOTE: This is an I/O type question, you need to write the whole code for taking input and printing the output.

Input Format

    • A single integer n, representing the number of rows in the pattern.

Output Format

    • A "V" shaped pattern with n rows, as described.

Constraints

    • 1 <= n

Examples

Input:

1
Output:

v
Input:

2
Output:

\ /
 v
Input:

3
Output:

\   /
 \ /
  v
'''
n = int(input())

for r in range(1, n + 1):
    leading_spaces = r - 1

    if r == n:
        # Last row prints 'v'
        print(' ' * leading_spaces + 'v')
    else:
        # Other rows print '\' and '/' with varying spaces in between
        spaces_between_slashes = 2 * (n - r) - 1
        print(' ' * leading_spaces + '\\' + ' ' * spaces_between_slashes + '/')
```

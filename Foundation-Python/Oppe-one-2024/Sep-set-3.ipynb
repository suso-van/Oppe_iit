```python
def div_by_exactly_one(num: int, a: int, b: int) -> bool:
    '''Check if num is divisible by exactly one of a or b.

    Args:
        num, a, b : int - input numbers

    Returns:
        bool - True if num is divisible by exactly one of a or b, otherwise False.
    '''

    return (num % a == 0) ^ (num % b == 0)

# Example usage:
print(f"10 divided by exactly one of 2 or 5: {div_by_exactly_one(10, 2, 5)}") # Expected: False (divisible by both)
print(f"9 divided by exactly one of 3 or 2: {div_by_exactly_one(9, 3, 2)}")   # Expected: True (divisible by 3 only)
print(f"7 divided by exactly one of 7 or 3: {div_by_exactly_one(7, 7, 3)}")   # Expected: True (divisible by 7 only)
print(f"6 divided by exactly one of 4 or 5: {div_by_exactly_one(6, 4, 5)}")   # Expected: False (divisible by neither)
print(f"15 divided by exactly one of 3 or 5: {div_by_exactly_one(15, 3, 5)}") # Expected: False (divisible by both)
```

```python
def mm_dd_yy_to_yy_dd_mm(date: str) -> str:
    """
    Convert a date string from the format mm-dd-yy to yy-dd-mm.

    Args:
        date (str): The date in the format "mm-dd-yy".

    Returns:
        str: The date in the format "yy-dd-mm".

    Example:
        >>> mm_dd_yy_to_yy_dd_mm("12-25-21")
        "21-25-12"
    """


    mm, dd, yy = date.split('-')
    return f"{yy}-{dd}-{mm}"

print(mm_dd_yy_to_yy_dd_mm("12-25-21"))
```

```python
def increment_value_with_max_limit(d: dict, key: str, inc: int, limit: int):
    '''
    Increment the value for the given key in the
    dictionary by `inc`, but not beyond `limit`.

    Modify the value in the dictionary do not return the new value.

    Args:
        d : dict - dictionary with integer values
        key : str - key to increment in the dictionary
        inc : int - increment value
        limit : int - maximum limit for the value

    Returns:
        None
    '''

    d[key] = min(d[key]+inc, limit)

# Example Usage:
my_dict = {'score': 10, 'level': 5}
print(f"Initial dictionary: {my_dict}")

# Increment 'score' by 5 with a limit of 12
increment_value_with_max_limit(my_dict, 'score', 5, 12)
print(f"After incrementing 'score' (limit 12): {my_dict}") # Expected: {'score': 12, 'level': 5}

# Increment 'level' by 2 with a limit of 10
increment_value_with_max_limit(my_dict, 'level', 2, 10)
print(f"After incrementing 'level' (limit 10): {my_dict}") # Expected: {'score': 12, 'level': 7}

# Try to increment 'score' beyond its limit, it should cap at 12
increment_value_with_max_limit(my_dict, 'score', 5, 12)
print(f"After trying to increment 'score' beyond limit: {my_dict}") # Expected: {'score': 12, 'level': 7}

# Try to increment a non-existent key, this will raise a KeyError unless handled
# For this example, we assume key exists as per docstring implicitly.
# d[key] will cause KeyError if 'new_key' doesn't exist. Let's make it exist for demonstration.
my_dict['new_key'] = 0
increment_value_with_max_limit(my_dict, 'new_key', 3, 5)
print(f"After incrementing 'new_key': {my_dict}") # Expected: {'score': 12, 'level': 7, 'new_key': 3}
```

```python
def last_word_starts_with_upper_case(sentence: str):
    """
    Find the last word in a sentence that starts with an uppercase letter.

    Args:
        sentence (str): The input sentence.

    Returns:
        str or None: The last word starting with an uppercase letter, or None if no such word exists.
    """


    last_word = None
    for word in sentence.split():
        if word[0].isupper():
            last_word = word
    return last_word

# Example Usage
print(last_word_starts_with_upper_case("Hello World"))
print(last_word_starts_with_upper_case("a small brown Fox"))
print(last_word_starts_with_upper_case("no uppercase words here"))
print(last_word_starts_with_upper_case("One Two Three"))
```

SECTION THREE PROBLEM-1

```python

def get_filter(criteria):
    if criteria == 'continuous':
        return lambda x: x[-3:].lower() == 'ing'
    if criteria == 'vowel_rich':
        return lambda x: len([c for c in x.lower() if c in 'aeiou'])>5
    if criteria == 'consonant_rich':
        return lambda x: len([c for c in x.lower() if c.isalpha() and c not in 'aeiou'])>5
    if criteria == 'sorted':
        return lambda x: "".join(sorted(x.lower())) == x.lower()

def get_words_by_criteria(words, criteria=None):
    """
    Filters words based on the given criteria.

    Args:
        words (list): List of words.
        criteria (str): The filter criteria.

    Returns:
        list: Filtered list of words matching the criteria, or None if invalid criteria.
    """


    filter_func = get_filter(criteria)
    if filter_func is not None:
        return list(filter(filter_func, words))

# Example usage to demonstrate the functions
words = ["singing", "dancing", "amazing", "rhythm", "aeiouaeiou", "apple", "banana", "hello", "billows", "ace", "bees", "act", "abhors", "cat", "bat", "art"]

print("Continuous:", get_words_by_criteria(words, 'continuous'))
print("Vowel-rich:", get_words_by_criteria(words, 'vowel_rich'))
print("Consonant-rich:", get_words_by_criteria(words, 'consonant_rich'))
print("Sorted:", get_words_by_criteria(words, 'sorted'))
print("Invalid criteria:", get_words_by_criteria(words, 'invalid'))
```
